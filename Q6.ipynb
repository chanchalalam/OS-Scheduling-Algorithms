```python
# Initialize the queues and their priority ranges
queue1 = []  # priority range 1-5
queue2 = []  # priority range 6-10
queue3 = []  # priority range 11-15

# Get the number of processes from the user
num_processes = int(input("Enter the number of processes: "))

# Get the priority and burst time of each process from the user and add it to the appropriate queue based on priority range
for i in range(num_processes):
    priority = int(input("Enter the priority of process {}: ".format(i+1)))
    burst_time = int(input("Enter the burst time of process {}: ".format(i+1)))
    if priority >= 1 and priority <= 5:
        queue1.append((i+1, priority, burst_time))
    elif priority >= 6 and priority <= 10:
        queue2.append((i+1, priority, burst_time))
    elif priority >= 11 and priority <= 15:
        queue3.append((i+1, priority, burst_time))
    else:
        print("Invalid priority range. Process not added to any queue.")

# Initialize time and quantum time for each queue
time = 0
quantum_time = 10

# Initialize the waiting time list
waiting_time = [0] * num_processes

# Print the table header
print("{:<10} {:<10} {:<10} {:<10} {:<10}".format("Process", "Priority", "Queue", "Burst Time", "Waiting Time"))
print("-" * 60)

# Execute the processes in each queue according to their scheduling algorithm
while queue1 or queue2 or queue3:
    # Execute processes in queue 1 using Round Robin scheduling with quantum time 4
    if queue1:
        process = queue1.pop(0)
        if process[2] > quantum_time:
            process = (process[0], process[1], process[2]-quantum_time)
            queue1.append(process)
            time += quantum_time
            waiting_time[process[0]-1] += quantum_time
            print("{:<10} {:<10} {:<10} {:<10} {:<10}".format(process[0], process[1], "1", quantum_time, waiting_time[process[0]-1]))
        else:
            time += process[2]
            waiting_time[process[0]-1] += process[2]
            print("{:<10} {:<10} {:<10} {:<10} {:<10}".format(process[0], process[1], "1", process[2], waiting_time[process[0]-1]))
    
    # Execute processes in queue 2 using Priority scheduling
    elif queue2:
        queue2.sort(key=lambda x: x[1], reverse=True)  # sort by descending order of priority
        process = queue2.pop(0)
        time += process[2]
        waiting_time[process[0]-1] += time - process[2]
        print("{:<10} {:<10} {:<10} {:<10} {:<10}".format(process[0], process[1], "2", process[2], waiting_time[process[0]-1]))
    
    # Execute processes in queue 3 using First Come First Serve scheduling
    elif queue3:
        process = queue3.pop(0)
        time += process[2]
        waiting_time[process[0]-1] += time - process[2]
        print("{:<10} {:<10} {:<10} {:<10} {:<10}".format(process[0], process[1], "3", process[2], waiting_time[process[0]-1]))

# Calculate the average waiting time
avg_waiting_time = sum(waiting_time) / num_processes

# Print the average waiting time
print("-" * 60)
print("Average waiting time: {:.2f}".format(avg_waiting_time))

```
